```python
def count_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Đếm số lượng outlier
    num_outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)].shape[0]
    return num_outliers
outliers_income = count_outliers_iqr(data_1, 'Thu nhập')
outliers_expense = count_outliers_iqr(data_1, 'Chi tiêu (trung bình một tháng)')
outliers_hoadondien = count_outliers_iqr(data_1, 'Hóa đơn tiền điện')

print(f"Số người có thu nhập bất thường (outlier): {outliers_income}")
print(f"Số người có chi tiêu bất thường (outlier): {outliers_expense}")
print(f"Số người có hóa đơn điện bất thường (outlier): {outliers_hoadondien}")
```

```python
def remove_outliers_iqr(data, column):
    # Tính Q1 (25th percentile), Q3 (75th percentile), và IQR
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    # Tính lower và upper bound
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Lọc dữ liệu để chỉ giữ lại các giá trị trong phạm vi hợp lý (không phải outliers)
    cleaned_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return cleaned_data
# Áp dụng hàm remove_outliers_iqr để loại bỏ outliers trong cột 'Thu nhập' và 'Chi tiêu'
data_1= remove_outliers_iqr(data_1, 'Thu nhập')
data_1= remove_outliers_iqr(data_1, 'Chi tiêu (trung bình một tháng)')
data_1= remove_outliers_iqr(data_1, 'Hóa đơn tiền điện')
data_1.describe()
```

```python
# Điền giá trị thiếu cho từng cột
# Biến phân loại: Điền giá trị thiếu bằng mode (giá trị xuất hiện nhiều nhất).
# Biến liên tục: Điền giá trị thiếu bằng median (trung vị).

# Với cột 'Trình độ học vấn': điền bằng giá trị xuất hiện nhiều nhất (mode)
data_1['Trình độ học vấn'] = data_1['Trình độ học vấn'].fillna(data_1['Trình độ học vấn'].mode()[0])

# Với cột 'Tình trạng hôn nhân': điền bằng mode
data_1['Tình trạng hôn nhân'] = data_1['Tình trạng hôn nhân'].fillna(data_1['Tình trạng hôn nhân'].mode()[0])


# Với cột 'Tình trạng công việc hiện tại (số năm kinh nghiệm)': điền bằng median
data_1['Tình trạng công việc hiện tại (số năm kinh nghiệm)'] = data_1['Tình trạng công việc hiện tại (số năm kinh nghiệm)'].fillna(data_1['Tình trạng công việc hiện tại (số năm kinh nghiệm)'].median())
```

```python
# Tính ma trận tương quan
corr_matrix = df.corr()

# Vẽ heatmap tương quan
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Tương Quan Giữa Các Thuộc Tính")
# plt.show()
```

```python
# Lọc ra các cặp cột có tương quan cao (giá trị tuyệt đối của hệ số tương quan lớn hơn 0.8)
threshold = 0.6

high_correlation = []

# Duyệt qua ma trận tương quan và lưu các cặp có tương quan cao
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            high_correlation.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j]))

# In ra các cặp cột có tương quan cao
for col1, col2, corr_value in high_correlation:
    print(f"Các cột {col1} và {col2} có hệ số tương quan là: {corr_value:.2f}")
```

```python
# Đặt ngưỡng tương quan
corr_threshold = 0.35

# Cột label
label_col = 'Kha nang tra no'

# Danh sách các cột có tương quan cao với label, cùng với giá trị tương quan
high_corr_with_label = []

# Lặp qua tất cả các cột để kiểm tra tương quan với cột label
for col in corr_matrix.columns:
    if col != label_col:
        corr_value = corr_matrix.loc[label_col, col]
        if abs(corr_value) > corr_threshold:
            high_corr_with_label.append((col, corr_value))

# Sắp xếp theo độ tương quan giảm dần (tuyệt đối)
high_corr_with_label = sorted(high_corr_with_label, key=lambda x: abs(x[1]), reverse=True)

# In ra kết quả
print(f"\nCác thuộc tính có tương quan cao với '{label_col}' (>|{corr_threshold}|), sắp xếp giảm dần:")
for col, value in high_corr_with_label:
    print(f"{label_col} ↔ {col}: {value:.2f}")
```

```python
from sklearn.metrics import roc_curve, auc
def _plot_roc_curve(fpr, tpr, thres, auc):
    plt.figure(figsize = (10, 8))
    plt.plot(fpr, tpr, color='darkorange', lw=2, linestyle='--', label='ROC curve (area = %0.2f)'%auc)
    plt.plot([0, 1], [0, 1], '--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate', color="r")
    plt.ylabel('True Positive Rate', color="g")
    plt.legend(loc='lower right')
    plt.title('ROC Curve')
```

```python
# Define the models to evaluate
models = [
    ('LR', LogisticRegression()),
    ('DT', DecisionTreeClassifier()),
    ('RF', RandomForestClassifier()),
    
]

# Evaluate each model on the test set using 10-fold cross-validation and store the results
results = []
names = []
for name, model in models:
    cv_results = cross_val_score(model, X_train, y_train, cv=10, scoring='accuracy')
    results.append(cv_results)
    names.append(name)

# Generate a boxplot comparing the model performance
fig = plt.figure()
fig.suptitle('Machine Learning Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.show()
```
