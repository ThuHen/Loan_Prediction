1. Import dataset and libraries

```python
# Import dataset and libraries
import pandas as pd # thư viện làm việc với data frame
import seaborn as sns # vẽ hình
import numpy as np # thư viện làm việc với con số
import matplotlib.pyplot as plt # vẽ hình
data = pd.read_csv("Data_credit.csv", header=0, encoding="utf-8")
```

```python
from copy import deepcopy
data_0 = deepcopy(data)
#data_0.head(5)
data_0.head(1)
```

2. Viewing number and types of features in dataset

```python
# Kiểm tra các features (số lượng, kiểu dữ liệu)
data_0.info()
```

```python
# Xóa cột ID khách hàng vì không có ý nghĩa
data_1 = data_0.drop(columns=['ID khách hàng'])
```

2. Handle missing values

```python
# Kiểm tra các biến có tồn tại missing data hay không
data_1.isnull().any()
```

```python
# Kiểm tra số lượng giá trị thiếu trong mỗi cột
missing_values = data_1.isnull().sum()

# Chỉ giữ lại các cột có giá trị thiếu (missing data)
missing_values = missing_values[missing_values > 0]

# Tính tỷ lệ phần trăm giá trị thiếu cho mỗi cột
missing_percentage = (missing_values / len(data_1)) * 100

# Kết hợp số lượng và tỷ lệ phần trăm vào một DataFrame
missing_data = pd.DataFrame({
    'Total Missing': missing_values,
    'Percentage Missing': missing_percentage
})

# In ra kết quả
print("\nSố lượng và tỷ lệ phần trăm giá trị thiếu của từng cột:")
print(missing_data)
```

```python
# Điền giá trị thiếu cho từng cột

# Với cột 'Trình độ học vấn': điền bằng giá trị xuất hiện nhiều nhất (mode)
data_1['Trình độ học vấn'] = data_1['Trình độ học vấn'].fillna(data_1['Trình độ học vấn'].mode()[0])

# Với cột 'Tình trạng hôn nhân': điền bằng mode
data_1['Tình trạng hôn nhân'] = data_1['Tình trạng hôn nhân'].fillna(data_1['Tình trạng hôn nhân'].mode()[0])

# Với cột 'Hóa đơn tiền điện': điền bằng giá trị trung vị (median)
data_1['Hóa đơn tiền điện'] = data_1['Hóa đơn tiền điện'].fillna(data_1['Hóa đơn tiền điện'].median())

# Với cột 'Tình trạng công việc hiện tại (số năm kinh nghiệm)': điền bằng median
data_1['Tình trạng công việc hiện tại (số năm kinh nghiệm)'] = data_1['Tình trạng công việc hiện tại (số năm kinh nghiệm)'].fillna(data_1['Tình trạng công việc hiện tại (số năm kinh nghiệm)'].median())
```

```python
# Kiểm tra lại còn tồn tại missing value hay không
data_1.isnull().sum()
```

Loại bỏ trùng lặp

```python
# Check số dòng bị trùng trong tập dữ liệu
data_1.duplicated().sum()
```

```python
# Xem các dữ liệu bị trùng lặp
duplicate = data_1[data_1.duplicated()]
print("Duplicate Rows:")
duplicate
```

```python
# Xóa các dòng bị trùng nhưng giữ lại dòng đầu tiên
data_2 = data_1.drop_duplicates(keep='first')
# Kiểm tra lại thông tin của các features (số lượng, kiểu dữ liệu)
data_2.info()
```

```python
# Đổi tên DataFrame
df = pd.DataFrame(data_2)
```

```python
# Convert data type float->int
# 23.7 → '23.7' → ['23', '7'] → '23'
for i in df.columns:
    df[i] = df[i].apply(lambda x: str(x).split('.')[0])
    df[i] = df[i].apply(lambda x: int(x))
df.head(5)
```

```python
# Đổi tên các biến sang không dấu
df=df.rename(columns = {'Khả năng trả nợ':'Kha nang tra no','Tình trạng công việc hiện tại (số năm kinh nghiệm)':'Tinh trang cong viec',
                        'Thu nhập':'Thu nhap','Tài sản thế chấp':'Tai san the chap',
                        'Giới tính':'Gioi tinh','Độ tuổi':'Do tuoi',
                        'Tình trạng hôn nhân':'Tinh trang hon nhan','Mục đích vay':'Muc dich vay',
                        'Số tiền dự kiến vay':'So tien vay','Thời hạn khoản vay':'Thoi han vay',
                        'Hóa đơn tiền điện':'Hoa don dien','Tình trạng sở hữu nhà ở':'Tinh trang so huu nha',
                        'Chi tiêu (trung bình một tháng)': 'Chi tieu', 'Chứng minh thu nhập':'Chung minh thu nhap',
                        'Số thành viên phụ thuộc trong gia đình': 'So nguoi phu thuoc', 'Trình độ học vấn':'Trinh do hoc van'
                       })
df.head(5)
```

```python
# Tính ma trận tương quan
corr_matrix = df.corr()

# Vẽ heatmap tương quan
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap Tương Quan Giữa Các Thuộc Tính")
plt.show()
```

```python
# Đặt ngưỡng tương quan
corr_threshold = 0.3

# Cột label
label_col = 'Kha nang tra no'

# Danh sách các cột có tương quan cao với label, cùng với giá trị tương quan
high_corr_with_label = []

# Lặp qua tất cả các cột để kiểm tra tương quan với cột label
for col in corr_matrix.columns:
    if col != label_col:
        corr_value = corr_matrix.loc[label_col, col]
        if abs(corr_value) > corr_threshold:
            high_corr_with_label.append((col, corr_value))

# Sắp xếp theo độ tương quan giảm dần (tuyệt đối)
high_corr_with_label = sorted(high_corr_with_label, key=lambda x: abs(x[1]), reverse=True)

# In ra kết quả
print(f"\nCác thuộc tính có tương quan cao với '{label_col}' (>|{corr_threshold}|), sắp xếp giảm dần:")
for col, value in high_corr_with_label:
    print(f"{label_col} ↔ {col}: {value:.2f}")
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Giả sử df là DataFrame của bạn, và 'Khả năng trả nợ' là cột nhãn (label)
X = df.drop(columns='Kha nang tra no')  # Các biến đầu vào
y = df['Kha nang tra no']               # Nhãn (label)

# Dùng Fisher Score (ANOVA F-test)
selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(X, y)

# Lấy tên và điểm F
fisher_scores = selector.scores_
feature_names = X.columns

# Kết hợp và sắp xếp
fisher_result = sorted(zip(feature_names, fisher_scores), key=lambda x: x[1], reverse=True)

# In kết quả
print("Fisher scores của các thuộc tính với 'Khả năng trả nợ':")
for feature, score in fisher_result:
    print(f"{feature}: {score:.2f}")
```

```python
target = ['Kha nang tra no']
features = ['Tai san the chap','Chung minh thu nhap','Tinh trang hon nhan','Thu nhap','Trinh do hoc van','Tinh trang so huu nha']
X = df[features]
y = df[target]
```

 Train Test Split

```python
# train, test split
from sklearn.model_selection import train_test_split
n_state = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = n_state)
X_train.shape, X_test.shape
```

 kiểm tra mất cân bằng dữ liệu (imbalanced data)

```python
print(np.count_nonzero(y_train == 1))
print(len(y_train) - np.count_nonzero(y_train == 1))
```

chuan hoa dữ lieu

```python
# Using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Scale train and test set
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

LOGISTIC REGRESSION

```python
# Xây dựng mô hình Logistic regression
LR_classifier = LogisticRegression(random_state=n_state)

# Đưa giá trị X_train và y_train vào mô hình Logistic regression
# Huấn luyện mô hình (dùng .ravel() để đảm bảo y là 1D)
LR_classifier.fit(X_train, y_train.values.ravel())

# Dự báo giá trị biến phụ thuộc
y_pred = LR_classifier.predict(X_train)

print('Confusion matrix:')
print(pd.DataFrame(confusion_matrix(y_train,y_pred)),'\n')

print('Classification report:')
print(classification_report(y_train,y_pred))

print('Logistic Regression accuracy: ', round(accuracy_score(y_train, y_pred),4))
```
